```python
import yt_dlp

def sc_info(url):
    """
    Extract SoundCloud metadata only (no download).
    Works for single tracks and playlists.
    """
    ydl_opts = {
        "quiet": True,
        "skip_download": True,
        "noplaylist": False,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=False)

    return info
```

```python
import json

link = "https://soundcloud.com/jayyw2yz/sets/sad-aw-man"
info = sc_info(link)

print(json.dumps(info, indent=2, ensure_ascii=False))
```
